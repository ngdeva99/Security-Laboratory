```python
def encryption():
    plain=str(input("\nEnter the plain text : "))
    row=int(input("Enter no. of rows: "))
    col=int(input("Enter no. of columns: "))
    print("Enter the keys")
    key=[int(input()) for i in range(col)]
    encrypt=[]
    
    mat=[['0' for i in range(col)] for j in range(row)]
    k=0
    val=82
    for i in range(row):
        for j in range(col):
            if k>=len(plain):
                mat[i][j]=chr(val)
                val=val+1
            else :
                mat[i][j]=plain[k]
                k=k+1
            
        

    for i in range(len(key)):
        for j in range(row):
            if mat[j][key[i]]!='0':
                encrypt.append(mat[j][key[i]])

    print("Cipher text:"+"".join(encrypt))
        
            

def decryption():
    cipher=str(input("Enter the cipher text : "))
    row=int(input("Enter no. of rows: "))
    col=int(input("Enter no. of columns: "))

    print("Enter the keys")
    key=[int(input()) for i in range(col)]

    mat=[['0' for i in range(col)] for j in range(row)]
    k=0
    for i in range(len(key)):
        for j in range(row):
            if k<len(cipher):
                mat[j][key[i]]=cipher[k]
                k=k+1
    decrypt=[]
    for i in range(row):
        for j in range(col):
            decrypt.append(mat[i][j])

    
    print("Plain text:"+"".join(decrypt))


while True:
    choice=int(input("\n1. Encryption\n2. Decryption\n3. Exit "))
    if choice==1:
        encryption()
    elif choice==2:
        decryption()
    else :
        break
```
